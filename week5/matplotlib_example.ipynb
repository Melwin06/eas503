```python
import matplotlib.pyplot as plt

```

```python
def sim_coin_toss(number_of_simulations):
    import random
    output = {}
    for i in range(number_of_simulations):
        toss = random.randint(0,1)
        
        if toss == 0:
            toss = 'Head'
        else:
            toss = 'Tail'
            
        if toss not in output:
            output[toss] = 0
            
        output[toss] += 1
    
    return output
```

```python
ten_sims = sim_coin_toss(10)
```

```python
print(ten_sims)
```

```python
data = ten_sims
group_data = list(data.values())
group_names = list(data.keys())
```

```python
fig, ax = plt.subplots()
ax.bar(group_names, group_data)
ax.set(xlabel='Coin', ylabel='Frequency',
       title='Simulatation with 10 Coin Tosses')
```
